```python

import pandas as pd
from sodapy import Socrata
import passwords
from passwords import *

client = Socrata("data.ny.gov", API_TOKEN)
# results = client.get("wujg-7c2s", limit=10000)

# df = pd.DataFrame.from_records(results)

station_name = "Bedford Park Blvd-Lehman College (4)"
station_name2 = "8 St-NYU (R,W)"

filtered_data = []

# Set the chunk size and total rows you want to load (1 million in this case)
chunk_size = 10000
max_rows = 1000000
offset = 0
```

```python

# Loop to keep fetching data in chunks and filter for the station names
while offset < max_rows:
    # Fetch data from the API with pagination using the offset and limit
    results = client.get("wujg-7c2s", limit=chunk_size, offset=offset)
    
    # If no more results, break the loop
    if len(results) == 0:
        break
    
    # Load the results into a DataFrame
    df = pd.DataFrame.from_records(results)
    
    # Filter for the station names
    filtered_chunk = df[df['station_complex'].isin([station_name, station_name2])]
    
    # Append the filtered data to the list
    if not filtered_chunk.empty:
        filtered_data.append(filtered_chunk)
    
    # Update the offset to fetch the next chunk
    offset += chunk_size

# Concatenate the filtered data into a single DataFrame
if filtered_data:
    final_filtered_data = pd.concat(filtered_data, ignore_index=True)
    print(final_filtered_data)
else:
    print("No matching stations found in the loaded data.")
```
